```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
from sklearn.cluster import KMeans
```

```python
image= cv2.imread('scenery.jpeg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
print(image.shape)
```

```python
plt.imshow(image)
plt.show()
```

```python
X=image.reshape(-1,3)
print(X.shape)
```

```python
k=10
model=KMeans(n_clusters=4)
model.fit(X)
```

```python
centroids=model.cluster_centers_
```

```python
print(centroids)
```

```python
colors = np.array(centroids, dtype='uint8')
```

```python
print(colors)
```

```python
i=1
for color in colors:
    plt.subplot(1,k,i)
    plt.axis('off')
    i=i+1
    mat=np.zeros((100,100,3),dtype='uint8')
    mat[:,:,:]=color
    plt.imshow(mat)
plt.show()
```
